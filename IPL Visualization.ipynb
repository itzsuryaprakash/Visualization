```python
sorted_df = df.sort_values('total_runs', ascending=False) #sorting the top 10 players from whole dataset
```

```python
top_10_players = sorted_df.head(10)
plt.bar(top_10_players['batsman'], top_10_players['total_runs']) #diplaying it in barchart

plt.xlabel('Player')
plt.ylabel('Runs')
plt.title('Top 10 Players by Runs Scored')

plt.xticks(rotation=45)

plt.show()
```

```python
colors=['red','green','orange','purple','cyan','yellow','lightgreen','pink','brown','blue']

plt.pie(top_10_players['total_runs'], labels=top_10_players['batsman'], autopct='%1.1f%%',colors=colors) #displaying it in pie chart

plt.title('Top 10 Players by Runs Scored')
```

```python
batsman = ['V Kohli', 'MS Dhoni', 'SK Raina', 'AB de Villiers', 'RG Sharma']
filtered_df = df[df['batsman'].isin(batsman)]
```

```python
sorted_df = filtered_df.sort_values('total_runs', ascending=False)
```

```python
top_players = sorted_df.head(5)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_players['batsman'], top_players['total_runs'])
plt.xlabel('Players')
plt.ylabel('Runs')
plt.title('Top 5 Players by Runs')

plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_players['total_runs'], labels=top_players['batsman'], autopct='%1.1f%%')
plt.title('Top 5 Players by Runs')
```

```python
df1=pd.read_csv('matches.csv')
```

```python
df1 = df1.iloc[:,:-1]
df1.dropna(inplace=True)
```

```python
df1["team1"].unique() #finding the unique values in dataset
```

```python
#for Delhi Capitals
df1['team1']=df1['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
df1['team2']=df1['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
df1['winner']=df1['winner'].str.replace('Delhi Daredevils','Delhi Capitals')
#for sunrisers Hyderabad
df1['team1']=df1['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df1['team2']=df1['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df1['winner']=df1['winner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
plt.figure(figsize = (10,6))
sns.countplot(y = 'winner',data = df1,order= df1['winner'].value_counts().index)
plt.xlabel('Wins')
plt.ylabel('Team')
plt.title('Number of  IPL  matches won by each team')
```

```python
plt.figure(figsize = (10,6))
sns.countplot(y = 'venue',data = df1,order = df1['venue'].value_counts().iloc[:10].index)
plt.xlabel('No of matches',fontsize=12)
plt.ylabel('Venue',fontsize=12)
plt.title('Total Number of matches played in different stadium')
```
